```python
def Q1(One_plus, BBK):

# Finding mean of one plus
  xbar = 0
  for i in One_plus:
    xbar = xbar + i
  xbar = xbar/(len(One_plus))
# Finding mean of BBK
  ybar = 0
  for j in BBK:
    ybar = ybar + j
  ybar = ybar/(len(BBK))

# Finding sample and population sd of One plus
  sum = 0
  sum_sq = 0
  for i in One_plus:
    sum = sum + (i-xbar)
    sum_sq = sum_sq + ((i-xbar)*(i-xbar))
  psd = pow(sum_sq/len(One_plus), 0.5) #Answer 2
  psdxs = pow(sum_sq/(len(One_plus)-1),0.5)


# Finding sample and population sd of One plus
  sumy = 0
  sum_sqy = 0
  for i in BBK:
    sumy = sumy + (i-ybar)
    sum_sqy = sum_sqy + ((i-ybar)*(i-ybar))
  psdyp = pow(sum_sqy/len(BBK), 0.5)
  psdys = pow(sum_sqy/(len(BBK)-1),0.5) #Answer 3

#Finding sample covariance 
  sumforcov = 0
  for a in range(len(One_plus)):
    sumforcov += ((One_plus[a]- xbar)*(BBK[a]-ybar))
  cov = sumforcov/(len(One_plus)-1) #Answer 4

#Finding correlation coefficient 
  clc = cov/(psdys*psdxs)  #Answer 5

  return psd, psdys, cov, clc



a=[6,2,1,1,2,1,6]
b=[10,10,11,11,10,11,16]
c= list(Q1(a,b))
print(c)
```

```python
def Q2(total,row1,row2):
  dull = row1[2]+row2[2]
  P1= dull/total #Answer 9

  poor = 0
  for i in row2:
    poor += i 
  P2 = poor/total #Answer 10

  good = 0
  for i in row1:
    good += i

  bl = row1[3]
  P3 = bl/good #Answer 11

  bright = row1[0]+ row2[0]
  P4 = (row2[0]/bright)*100 #Answe 12

  avg = row1[1] + row2[1] 
  P5 = (row1[1]/avg)*100 #Answer 13

  return P1, P2, P3, P4, P5


a=[54,95,86,128]
b=[54,96,70,75]
q= list(Q2(658,a,b))
print(q)
```
